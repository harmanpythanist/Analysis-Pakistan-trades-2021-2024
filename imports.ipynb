# Jan 2024 imports vs Jan 2025 imports

```python
df_2024 = file[file['YEAR'] == 2024]
df_2024_jan = df_2024[df_2024['MONTH'] == 'January']

df_2025 = file[file['YEAR'] == 2025]
df_2025_jan = df_2025[df_2025['MONTH'] == 'January']
```

```python
# Sum Quantity, Rupees, Dollars for each category in January 2024 and 2025
summary_2024_jan = df_2024_jan.groupby('CATEGORY')[['QUANTITY', 'RUPEES', 'DOLLARS']].sum()
summary_2025_jan = df_2025_jan.groupby('CATEGORY')[['QUANTITY', 'RUPEES', 'DOLLARS']].sum()
```

```python
# Merge data of 2024 and 2025
comparison = summary_2024_jan.merge(summary_2025_jan, on='CATEGORY', suffixes=('_2024', '_2025'))

# new column = calculate difference
comparison['QUANTITY_DIFF'] = comparison['QUANTITY_2025'] - comparison['QUANTITY_2024']
comparison['RUPEES_DIFF'] = comparison['RUPEES_2025'] - comparison['RUPEES_2024']
comparison['DOLLARS_DIFF'] = comparison['DOLLARS_2025'] - comparison['DOLLARS_2024']

comparison
```

```python
comparison['DOLLARS_DIFF'].plot(kind='bar', figsize=(8,5), color='skyblue')
plt.title("Imports: Jan 2024 VS jan 2025")
plt.ylabel("Difference in import value (in dollars)")
plt.xlabel("Category")
plt.show()
```

```python
# For Vegetables
veg_2024 = df_2024_jan[df_2024_jan['PRODUCT'].str.contains('VEGETABLE', case=False)]
veg_2025 = df_2025_jan[df_2025_jan['PRODUCT'].str.contains('VEGETABLE', case=False)]

# For Fruits
fruits_2024 = df_2024_jan[df_2024_jan['PRODUCT'].str.contains('FRUIT', case=False)]
fruits_2025 = df_2025_jan[df_2025_jan['PRODUCT'].str.contains('FRUIT', case=False)]

# For Sports
sports_2024 = df_2024_jan[df_2024_jan['PRODUCT'].str.contains('SPORT', case=False)]
sports_2025 = df_2025_jan[df_2025_jan['PRODUCT'].str.contains('SPORT', case=False)]
```

# Most imports in which year (2021 - 2024)

```python
df_2024 = file[file['YEAR'] == 2024]
df_2023 = file[file['YEAR'] == 2023]
df_2022 = file[file['YEAR'] == 2022]
df_2021 = file[file['YEAR'] == 2021]

total_sales_year = [df_2024, df_2023, df_2022, df_2021]
year = ['2024', '2023', '2022', '2021']
```

```python
data_imports = []
for i,j in enumerate(total_sales_year):
    total_dollar = j['DOLLARS'].sum()
    data_imports.append(total_dollar)
    print(year[i], 'imports value (DOLLARS):', total_dollar)

```

```python
plt.plot(year, data_imports)
plt.bar(year, data_imports)
plt.scatter(year, data_imports)

plt.title("Each Year Imports revenue data")
plt.xlabel("Years")
plt.ylabel('Import value (dollar)')
plt.show()
```

# Food imports data in detail

```python
food_data = file[file['CATEGORY'] == 'FOOD GROUP']
food_january = food_data[food_data['MONTH'] == 'January']
food_imports_by_year = food_january.groupby('YEAR')['DOLLARS'].sum().sort_index()

print("Total Food imports value (in dollars) in January of each year:\n")
print(food_imports_by_year)
```

```python
plt.figure(figsize=(10,5))
plt.plot(food_imports_by_year.index, food_imports_by_year.values)
plt.scatter(food_imports_by_year.index, food_imports_by_year.values)
plt.title("Food import in january from 2021 to 2024")
plt.xlabel("Year")
plt.ylabel("Imports value in dollars")
plt.xticks(food_imports_by_year.index) 
plt.grid(True)
plt.show()
```

```python
food_2024 = file[(file['CATEGORY'] == 'FOOD GROUP') & (file['YEAR'] == 2024)]
food_imports_monthly = food_2024.groupby('MONTH')['DOLLARS'].sum()

month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
food_imports_monthly = food_imports_monthly.reindex(month_order)

print(food_imports_monthly)
```

```python
plt.figure(figsize=(10,5))
plt.plot(food_imports_monthly.index, food_imports_monthly.values, color='orange', linewidth=2)
plt.scatter(food_imports_monthly.index, food_imports_monthly.values, color = 'orange')
plt.title("Food imports of each month of 2024")
plt.xlabel("Months")
plt.ylabel("Imports value in dollars")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
food_data = file[file['CATEGORY'] == 'FOOD GROUP']

# Filter by year
food_2022 = food_data[food_data['YEAR'] == 2022]
food_2023 = food_data[food_data['YEAR'] == 2023]
food_2024 = food_data[food_data['YEAR'] == 2024]

month_order = ['January','February','March','April','May','June','July','August', 'September','October','November','December']

# Monthly sum
food_monthly_2022 = food_2022.groupby('MONTH')['DOLLARS'].sum().reindex(month_order)
food_monthly_2023 = food_2023.groupby('MONTH')['DOLLARS'].sum().reindex(month_order)
food_monthly_2024 = food_2024.groupby('MONTH')['DOLLARS'].sum().reindex(month_order)
```

```python
plt.figure(figsize=(12,6))

plt.plot(food_monthly_2022.index, food_monthly_2022.values, marker='o', label='2022', color='blue', linewidth=4)
plt.plot(food_monthly_2023.index, food_monthly_2023.values, marker='o', label='2023', color='green', linewidth=4)
plt.plot(food_monthly_2024.index, food_monthly_2024.values, marker='o', label='2024', color='orange', linewidth=4)

plt.title("Food imports for all months of 2022, 2023, 2024")
plt.xlabel("Month")
plt.ylabel("Imports value in dollars")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
years = [2021, 2022, 2023, 2024]
data = file[file['YEAR'].isin(years)]

month_order = ['January','February','March','April','May','June', 'July','August','September','October','November','December']
monthly_imports = (data.groupby(['YEAR', 'MONTH'])['DOLLARS'].sum().unstack().reindex(month_order, axis=1))
```

```python
monthly_imports_billion = monthly_imports / 1_000_000
```

```python
monthly_imports_billion
```

```python
plt.figure(figsize=(12,6))

plt.plot(month_order, monthly_imports_billion.loc[2021], marker='o', linewidth=4,label='2021')
plt.plot(month_order, monthly_imports_billion.loc[2022], marker='o', linewidth=4,label='2022')
plt.plot(month_order, monthly_imports_billion.loc[2023], marker='o', linewidth=4,label='2023')
plt.plot(month_order, monthly_imports_billion.loc[2024], marker='o', linewidth=4,label='2024')


plt.title("Monthly Imports (20212024)")
plt.xlabel("Month")
plt.ylabel("Imports value (Billion Dollars)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title="Year")
plt.tight_layout()
plt.show()

```
