# Jan 2024 imports vs Jan 2025 imports

```python
df_2024 = file[file['YEAR'] == 2024]
df_2024_jan = df_2024[df_2024['MONTH'] == 'January']
```

```python
df_2025 = file[file['YEAR'] == 2025]
df_2025_jan = df_2025[df_2025['MONTH'] == 'January']
```

```python
# Sum Quantity, Rupees, Dollars for each category in January 2024 and 2025
summary_2024_jan = df_2024_jan.groupby('CATEGORY')[['QUANTITY', 'RUPEES', 'DOLLARS']].sum()
summary_2025_jan = df_2025_jan.groupby('CATEGORY')[['QUANTITY', 'RUPEES', 'DOLLARS']].sum()
```

```python
# Merge data of 2024 and 2025
comparison = summary_2024_jan.merge(summary_2025_jan, on='CATEGORY', suffixes=('_2024', '_2025'))

# new column = Calculate difference
comparison['QUANTITY_DIFF'] = comparison['QUANTITY_2025'] - comparison['QUANTITY_2024']
comparison['RUPEES_DIFF'] = comparison['RUPEES_2025'] - comparison['RUPEES_2024']
comparison['DOLLARS_DIFF'] = comparison['DOLLARS_2025'] - comparison['DOLLARS_2024']

comparison
```

```python
comparison['DOLLARS_DIFF'].plot(kind='bar', figsize=(8,5), color='skyblue')
plt.title("Exports: Jan 2024 VS jan 2025")
plt.ylabel("Difference in exports value (in dollars)")
plt.xlabel("Category")
plt.show()
```

```python
# For Vegetables
veg_2024 = df_2024_jan[df_2024_jan['PRODUCT'].str.contains('VEGETABLE', case=False)]
veg_2025 = df_2025_jan[df_2025_jan['PRODUCT'].str.contains('VEGETABLE', case=False)]

# For Fruits
fruits_2024 = df_2024_jan[df_2024_jan['PRODUCT'].str.contains('FRUIT', case=False)]
fruits_2025 = df_2025_jan[df_2025_jan['PRODUCT'].str.contains('FRUIT', case=False)]

# For Sports
sports_2024 = df_2024_jan[df_2024_jan['PRODUCT'].str.contains('SPORT', case=False)]
sports_2025 = df_2025_jan[df_2025_jan['PRODUCT'].str.contains('SPORT', case=False)]
```

# Most imports in which year (2021 - 2024)

```python
years = [2021, 2022, 2023, 2024]
data = file[file['YEAR'].isin(years)]
```

```python
plt.figure(figsize=(12,6))

plt.plot(month_order, monthly_exports_billion.loc[2021], marker='o', linewidth=4,label='2021')
plt.plot(month_order, monthly_exports_billion.loc[2022], marker='o', linewidth=4,label='2022')
plt.plot(month_order, monthly_exports_billion.loc[2023], marker='o', linewidth=4,label='2023')
plt.plot(month_order, monthly_exports_billion.loc[2024], marker='o', linewidth=4,label='2024')


plt.title("Monthly exports (20212024)")
plt.xlabel("Month")
plt.ylabel("exports value (Billion Dollars)")

plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title="Year")
plt.tight_layout()
plt.show()

```
